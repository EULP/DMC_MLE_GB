```python
red = setup(data=df,target="default",session_id=123,normalize=True,categorical_features=["vehiculo_propio","historia_credito","estado_civil"],fix_imbalance=True)
```

```python
#2. Compare_models(): Comparar multiples modelos y los rankear
best_model = compare_models(sort="AUC")
best_model
```

```python
svm_model = create_model('svm') 
evaluate_model(svm_model)
```

```python
lr_model = create_model('lr') 
evaluate_model(lr_model)
```

```python
tuned_model_svm = tune_model(svm_model, optimize='AUC')
tuned_model_lr = tune_model(lr_model, optimize='AUC')
display(tuned_model_svm)
display(tuned_model_lr)
```

```python
evaluate_model(tuned_model_svm)
evaluate_model(tuned_model_lr)
```

```python
preds_svm = predict_model(tuned_model_svm)
preds_svm.head()
```

```python
preds_lr = predict_model(tuned_model_lr)
preds_lr.head()
```

```python
save_model(tuned_model_lr,"modelo_precio_final")
```
